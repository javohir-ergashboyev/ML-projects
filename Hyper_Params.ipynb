```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data=load_iris()
dir(data)
```

```python
df=pd.DataFrame(data.data, columns=data.feature_names)
df['flower']=data.target
df['flower']=df['flower'].apply(lambda x: data.target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(df.drop('flower', axis=1), data.target, test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='rbf', C=30, gamma ='auto')
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
}, cv=5, return_train_score=False)
clf.fit(data.data, data.target)
```

```python
res=pd.DataFrame(clf.cv_results_)
res.head()
```

```python
res[['param_C','param_kernel','mean_test_score']].sort_values(by='mean_test_score', ascending=False)
```

```python
clf.best_score_, clf.best_params_
```

# Best Model Selection

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
model_params={
    'svm':{
        'model': SVC(gamma='auto'),
        'params':{
            'kernel':['rbf', 'linear'],
            'C':[1,10,20]
        }
    },
    'logR':{
         'model': LogisticRegression(solver='liblinear', multi_class='auto'),
         'params':{
            'C':[1,5,10]
         }
    },
    'rf':{
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    }
}
```

```python
scores=[]

for model_name, mp in model_params.items():
  clf=GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
  clf.fit(data.data, data.target)
  scores.append({
      'model':model_name,
      'scores':clf.best_score_,
      'best_params':clf.best_params_
  })


scores
```
