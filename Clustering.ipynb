```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
np.random.seed(0)
centroids=[[3,3],[-3,-2],[2,-3],[0,0]]
X,y=make_blobs(n_samples=5000, centers=centroids, cluster_std=0.8)
```

```python
plt.scatter(X[:,0],X[:,1], marker='.')
plt.show()
```

```python
k_m=KMeans(init='random', n_clusters=4, n_init=15)
k_m.fit(X)
```

```python
k_m.labels_
```

```python
k_m.cluster_centers_
```

```python
# prompt: draw scatter for each cluster in different color

colors = ['red', 'green', 'blue', 'yellow']
for i in range(k_m.n_clusters):
    plt.scatter(X[k_m.labels_ == i, 0], X[k_m.labels_ == i, 1], c=colors[i], label=f'Cluster {i+1}')
plt.legend()
plt.show()

```
